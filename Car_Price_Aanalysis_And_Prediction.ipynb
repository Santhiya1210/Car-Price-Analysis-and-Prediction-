## Stage 2 â Data Cleaning and Pre-processing



```python
import warnings
warnings.filterwarnings('ignore')
```

# **Handle missing values**

```python
df['engV'].fillna(df['engV'].mean(),inplace=True)
```

```python
df['drive'].fillna(df['drive'].mode()[0],inplace=True)
```

```python
df['engV'] = df['engV'].round(2)
```

```python
df.isnull().sum()
```

# **Handle duplicates**

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

# **Derived Fields**

```python
df['drive_num'] = df['drive'].map({
    'front': 0,
    'rear': 1,
    'full': 2
})
```

```python
df['car_age'] = 2025 - df['year']
```

# **Skewness Check Before Outlier Treatment**

```python
df.skew(numeric_only=True)
```

# **Outlier Treatment**

```python
df = df[df['price'] > 0]
df = df[df['mileage'] < 500]
df = df[df['engV'] < 10]
df = df[df['mileage'] > 0]
df = df[df['engV'] > 0.5]
```

* I filtered the data to remove unrealistic or extreme values so that the dataset is clean for analysis

```python
df.describe()
```

# **Skewness reduction**

```python
df['price'] = np.log1p(df['price']) # to Reduce Skweness
```

```python
df.skew(numeric_only=True)
```

The skewness of the numerical columns was checked to understand the data distribution.

* Price â 0.05 â Almost normal (log transform applied, skewness reduced).

* Mileage â 0.67 â Slightly positively skewed, most cars have low mileage.

* Engine Volume (engV) â 1.80 â Moderately positively skewed, few cars have large engines.

* Year â -1.56 â Negatively skewed, most cars are recent.

* Drive_num â 0.78 â Mildly skewed, distribution is acceptable.

* Car_age â 1.56 â Moderately positively skewed, some older cars exist.
